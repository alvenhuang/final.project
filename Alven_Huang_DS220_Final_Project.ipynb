**Quick Background about dataset:**
1. ***date***: The day the game was played on
2. ***league_id***: Keeps track of which World Cup it was
3. ***league***: Name of the tournament
4. ***team_1***: The team considered the home team
5. ***team_2***: The team considered the away tam
6. ***spi_1***: "Soccer Power Index", a rating system for each team (updates consistently). This is for team_1.
7. ***spi_2***: "Soccer Power Index", a rating system for each team (updates consistently). This is for team_2.
8. ***prob_1***: The probabilty for team one to win
9. ***prob_2***: The probability for team two to win
10. ***probtie***: The probabilty for the teams to tie
11. ***proj_score1***: The projected score for team one
12. ***proj_score2***: The projected score for team two
13. ***score_1***: The actual score for team one
14. ***score_2***: The actual score for team two
15. ***xg1***: The amount of expected goals for team one
16. ***xg2***: The amount of expected goals for team two
17. ***nsxg1***: The amount of non shot expected goals for team one
18. ***nsxg2***: The amount of non shot expected goals for team two
19. ***adj_score1***: Adjusted score for team one based off performance
20. ***adj_score2***: Adjusted score for team two based off performance

**An example plot:** Here I decided to graph the total goal scored by each country on a bar graph. First, I created a total goals variable to count the total number of goals each team scored in the tournament. I used functions such as groupby, .sum(), and sort_values. Then I used my new variable with .plot to make it into a plot. I used the features 'kind', 'figsize', and 'color' to make my plot more appealing. Then I focused on my x and y labels and title. .Show() ultimately displays the plot.

```python
import matplotlib.pyplot as plt
total_goals = (soccer_data.groupby('team1')['score1'].sum() + soccer_data.groupby('team2')['score2'].sum()).sort_values(ascending=False)
total_goals.plot(kind='bar', figsize=(12, 6), color = 'maroon')
plt.title('Total Goals Scored by Each Country at the 2022 Qatar World Cup')
plt.xlabel('Country')
plt.ylabel('Total Goals')
plt.tight_layout()
plt.show()
```

**Plot Above:** The plot above has contains all the countries along with how many goals they scored in the tournament. This graph goes in descending order by goals scored.

**Answer:** The matches that contain 5 or more goals are

**Question 5: The Top 5 teams with the highest total expected goals throughout the World Cup.** I created a new variable called high_xg to add every team's xg in each game together. Then, I sort it by descending order, and only search for the top 5 highest teams.

---

**Answer:** The 5 teams with the Highest xg's are Argentina with 7.66, France with 7.66, Brazil with 5.90, England with 4.95, and Portugal with 4.65

```python
high_xg = soccer_data.groupby('team1')['xg1'].sum().sort_values(ascending=False).head(5)
print("Top 5 Teams With the Highest Expected Goals Scored at the 2022 Qatar World Cup:")
print(high_xg)

plt.figure(figsize=(10, 6))
high_xg.plot(kind='bar', color='skyblue')
plt.title('Top 5 Teams With the Highest Expected Goals at the 2022 Qatar World Cup')
plt.xlabel('Teams')
plt.ylabel('Expected Goals (xG)')
plt.show()
```

**Graph Above:** This graph above has the Teams as the x-axis and expected goals as the y-axis. Graph is for the teams with the highest expected goals at the 2022 World Cup

---



**Question 6: The Top 5 teams with the lowest total expected goals throughout the World Cup.** I created a new variable called low_xg to add every team's xg in each game together. Then, I sort it by descending order, and only search for the top 5 lowest teams.

---

**Answer:** The 5 teams with the lowest xg's are USA with 0.33, Mexico with 0.45, Australia with 0.47, Uruguay with 0.48, and Ecuador with 0.61.

```python
low_xg = soccer_data.groupby('team1')['xg1'].sum().sort_values(ascending=True).head(5)
print("Top 5 Teams With the Lowest Expected Goals Scored at the 2022 Qatar World Cup:")
print(low_xg)

plt.figure(figsize=(10, 6))
low_xg.plot(kind='bar', color='red')
plt.title('Top 5 Teams With the Lowest Expected Goals at the 2022 Qatar World Cup')
plt.xlabel('Teams')
plt.ylabel('Expected Goals (xG)')
plt.show()
```

**Graph Above:** This graph above has the Teams as the x-axis and expected goals as the y-axis. Graph is for the teams with the lowest expected goals at the 2022 World Cup.
